```python
df_subms = pd.read_json('data/submissions2.json')
df_subms
```

```python
df_classes = df_subms['class_id'].drop_duplicates().reset_index(name="class_id")
df_classes = df_classes.drop("index", axis=1)
df_classes
```

```python
hits = pd.to_numeric(df_subms_target['hitPercentage'], downcast='integer', errors='coerce').reset_index()
hits_list = hits['hitPercentage'].tolist()
df_subms_target = df_subms_target.drop('hitPercentage', axis=1)
df_subms_target = df_subms_target.assign(hitPercentage= hits_list)
df_performance = df_subms_target.drop('hitPercentage', axis=1)
df_performance = df_performance.assign(hitPercentage = hits_list)
#pontuação máxima em cada questão por auluno
df_performance= df_performance.groupby(['user_id', 'question_id','class_id', 'listQuestions_id'])['hitPercentage'].max().reset_index()
#soma das notas por lista
df_performance= df_performance.groupby(['user_id','class_id', 'listQuestions_id'])['hitPercentage'].sum().reset_index()
```

```python
#número de questões por lista
df_listas_questoes = df_subms_target[['listQuestions_id', 'question_id', 'class_id', 'user_id']].copy()
df_listas_questoes=df_listas_questoes.drop_duplicates(subset=['listQuestions_id', 'question_id','user_id','class_id'])
df_listas_questoes=df_listas_questoes.drop_duplicates(subset=['listQuestions_id', 'question_id'])
df_listas_questoes = df_listas_questoes.groupby('listQuestions_id')['question_id'].count().reset_index()
df_listas_questoes.columns.values[1]='questQuant'
df_listas_questoes
```

```python
listas_questoes_list = df_listas_questoes['listQuestions_id'].tolist()
listas_quant_list = df_listas_questoes['questQuant'].tolist()
lenght = len(listas_questoes_list)
for i in np.arange(0,lenght):
    lista = listas_questoes_list[i]
    numQuest = listas_quant_list[i]
    for row in np.arange(len(df_performance.index)):
        if(df_performance.iloc[row]['listQuestions_id']==lista):
            df_performance.at[row, 'hitPercentage']=df_performance.iloc[row]['hitPercentage']/numQuest

```

```python
#PEGAR APENAS SUBMISSÕES DA LISTA 1
#4f3452ae-87d4-4a5f-9532-c7634da2cff0

df_subms_lista1 = df_subms_target.loc[df_subms_target['listQuestions_id']=='4f3452ae-87d4-4a5f-9532-c7634da2cff0']
df_subms_count_lista1 = df_subms_lista1.groupby('user_id')['_id'].count().reset_index()

df_errors_lista1 = df_subms_lista1.loc[df_subms_lista1['hitPercentage']==0] 
df_errors_lista1 = df_errors_lista1.groupby('user_id')['hitPercentage'].count().reset_index()

#get users list
df_total_users = df_subms_count_lista1.drop('_id', axis=1)
users_list = df_total_users['user_id'].tolist()
lenght = len(users_list)

#columns of df_errors_lista1
usersErrors_noZeros = df_errors_lista1['hitPercentage'].tolist()
users_noZeros = df_errors_lista1['user_id'].tolist()

errorsList = []

for i in np.arange(0, lenght):
    user = users_list[i]
    errorsCount = 0;
    found = False
    noZeros_index = -1
    for j in np.arange(len(users_noZeros)):
        user_noZero = users_noZeros[j]
        if(user_noZero==user):
            noZeros_index = j           
    if(noZeros_index>-1):
        errorsList.append(usersErrors_noZeros[noZeros_index])
    else:
        errorsList.append(0)
df_errors_lista1 = df_total_users.copy()
df_errors_lista1['totalErrors'] = errorsList
df_errors_lista1

```

```python
#submissões com total acerto
df_acertos_subms = df_subms_lista1.loc[df_subms_lista1['hitPercentage']>0]
df_acertos_subms2 = df_acertos_subms.groupby('user_id')['hitPercentage'].count().reset_index()

users_acertos_subms = df_acertos_subms2['user_id'].tolist()
num_acertos_subms = df_acertos_subms2['hitPercentage'].tolist()
num_acertosList = []
for i in np.arange(len(users_list)):
    user = users_list[i]
    found = False
    for j in np.arange(len(users_acertos_subms)):
        if(users_acertos_subms[j]==user):
            num_acertosList.append(num_acertos_subms[j])
            found=True
    if(found==False):    
        num_acertosList.append(0)

df_acertos_subms3 = df_total_users.copy()
df_acertos_subms3['algumAcerto'] = num_acertosList
df_acertos_subms3
```

```python
#quantidade de questões certas na lista
df_acertos_lista = df_acertos_subms.groupby(['user_id', 'question_id'])['hitPercentage'].max().reset_index()
df_acertos_lista2 = df_acertos_lista.groupby('user_id')['question_id'].count().reset_index()

users_acertos_lista = df_acertos_lista2['user_id'].tolist()
num_acertos_lista = df_acertos_lista2['question_id'].tolist()
num_acertos_listaList = []

for i in np.arange(len(users_list)):
    user = users_list[i]
    found = False
    for j in np.arange(len(users_acertos_lista)):
        if(users_acertos_lista[j]==user):
            num_acertos_listaList.append(num_acertos_lista[j])
            found=True
    if(found==False):
        num_acertos_listaList.append(0)

df_acertos_lista2 = df_total_users.copy()
df_acertos_lista2['questoes_certas'] = num_acertos_listaList

df_acertos_lista2
```

```python
df_acertos_100 = df_subms_lista1.copy()
df_acertos_100 = df_acertos_100.loc[df_acertos_100['hitPercentage']==100]
df_acertos_100 = df_acertos_100.drop_duplicates(subset=['user_id', 'question_id']).reset_index()
df_acertos_100 = df_acertos_100.groupby('user_id')['user_id'].count().reset_index(name='acertos_100')
df_acertos_100

acertos100_list = df_acertos_100['acertos_100'].tolist()
acertos100Users_list = df_acertos_100['user_id'].tolist()
acertos100_update = []

for i in np.arange(len(users_list)):
    user = users_list[i]
    found =False
    for j in np.arange(len(acertos100Users_list)):
        user2 = acertos100Users_list[j]
        if(user==user2):
            found=True
            acertos100_update.append(acertos100_list[j])
    if(found==False):
        acertos100_update.append(0)

df_acertos_100 = df_total_users.copy()
df_acertos_100['acertos_100'] = acertos100_update
df_acertos_100
```

```python
df_users_listas = df_subms_target.copy()
df_users_listas = df_users_listas.drop_duplicates(subset=['user_id', 'listQuestions_id']).reset_index()
df_users_listas = df_users_listas.groupby('user_id')['listQuestions_id'].count().reset_index()

desistente = []
users_desistencia = df_users_listas['user_id'].tolist()
listas_desistencia = df_users_listas['listQuestions_id'].tolist()

for i in np.arange(len(listas_desistencia)):
    if(listas_desistencia[i]<4):
        desistente.append(1)
    else:
        desistente.append(0)
        
df_users_listas['desistente'] = desistente

desistente_target = []
for i in np.arange(len(users_list)):
    user = users_list[i]
    found = False
    for j in np.arange(len(users_desistencia)):
        user2 = users_desistencia[j]        
        if(user==user2):
            desistente_target.append(desistente[j])
            found = True
    if(found==False):
         desistente_target.append(1)

df_users_desistentes = df_total_users.copy()
df_users_desistentes['desistente'] = desistente_target
df_users_desistentes
    
```

```python
df_subms_count = df_subms_lista1.copy()
df_subms_count = df_subms_count.groupby('user_id')['user_id'].count().reset_index(name='submissoes')
df_subms_count
```

```python
df_questoes_tentadas = df_subms_lista1.copy()
df_questoes_tentadas = df_questoes_tentadas.drop_duplicates(subset=['user_id', 'question_id']).reset_index()
df_questoes_tentadas = df_questoes_tentadas.groupby('user_id')['question_id'].count().reset_index(name = 'questoes_tentadas')
df_questoes_tentadas_list = df_questoes_tentadas['questoes_tentadas'].tolist()

df_subms_por_questao = df_subms_count.copy()
df_subms_por_questao['subms_por_questao'] = df_subms_count['submissoes']/df_questoes_tentadas['questoes_tentadas']
df_subms_por_questao
```

```python
df_totalTimeInSeconds = df_subms_lista1.copy()
df_totalTimeInSeconds = df_totalTimeInSeconds.groupby('user_id')['timeConsuming'].sum().reset_index(name='timeToFinishList')
df_totalTimeInSeconds['timeToFinishList']=df_totalTimeInSeconds['timeToFinishList'].divide(1000)
df_totalTimeInSeconds
```

```python
df_final = df_errors_lista1.copy()
df_final['algumAcerto'] = df_acertos_subms3['algumAcerto']
df_final['questoesCertas'] = df_acertos_lista2['questoes_certas']
df_final['acertos_100'] = df_acertos_100['acertos_100']
df_final['submissoes'] = df_subms_count['submissoes']
df_final['subms_por_questao'] = df_subms_por_questao['subms_por_questao']
df_final['timeToFinishList'] = df_totalTimeInSeconds['timeToFinishList']
df_final['percTotalError'] = df_errors_lista1['totalErrors']/df_subms_count['submissoes']
df_final['percAlgumAcerto'] = df_acertos_subms3['algumAcerto']/df_subms_count['submissoes']
df_final['percAcerto100'] = df_acertos_100['acertos_100']/df_subms_count['submissoes']
df_final
```

```python
#DECISION TREE
y_desist = df_users_desistentes['desistente'].tolist()
x_dados = df_final.drop('user_id', axis=1)
x_dados = x_dados.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_dados, y_desist, test_size=0.30, shuffle=True, random_state=123)

dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=10, random_state=1)
dt_model.fit(X_train, y_train)
```

```python
y_test_pred = dt_model.predict(X_test)

accuracy_score(y_test, y_test_pred)
```
